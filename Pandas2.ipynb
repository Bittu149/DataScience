```python
import numpy as np
import pandas as pd
```

```python
#Finding Missing Data
# Nan means Not a Number
data = {
'A':[1,2,np.nan,4,5],
'B':[1,2,3,4,5],
'C':[1,2,3,np.nan,np.nan],
'D':[1,np.nan,np.nan, np.nan,5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna() # It will show true where data is missing.
```

```python
df.isna().sum() # It will show how many missing value in each column.
```

```python
df.isna().any() # it will show true if any missing value is present in that column.
```

```python
df
```

```python
df.dropna() # it will drop all the rows which have any missing value in it.
```

```python
df
```

```python
df.dropna(thresh=3) # it will drop all the rows which have at least 3 non-missing values.
# Drop rows wise drop hota hai.
```

```python
#Filling the missing values 
df.fillna(0) # it will fill all the missing value with 0.
```

```python
values = {'A':0,'B':100,'C':300,'D':400}
df.fillna(value=values) # it will fill all the missing value with the values provided in the dictionary
# fillna se column wise fill hota hai.
```

```python
df
```

```python
df.fillna(df.mean()) # it will fill all the missing value with the mean of that column.
# it will work only with number columns.
```
