```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Category':['A','B','A','B','A','B','A','B'],
    'Store':['S1','S1','S2','S2','S1','S2','S2','S1'],
    'Sales':[100,200,150,250,120,180,200,300],
    'Quantity':[10,15,12,18,8,28,15,25],
    'Date':pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Group By category and calculate the sum of sales
df.groupby('Category')['Sales'].sum() # Group by category and calculate the sum of sales
# it create an object of series type
```

```python
df.groupby('Store')['Sales'].sum() 
```

```python
df.groupby(['Category','Store'])['Sales'].sum()
```

```python
#Aggregation basic aggregation function :- mean, median, mode, standard deviation, varience, min, max
df['Sales'].mean()
```

```python
df['Sales'].agg(['mean','min','max','count','median'])
```
